### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/u.user).

### Step 3. Assign it to a variable called users.

```python
column_names = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
users = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/u.user', sep='|', header=None, names=column_names, skiprows=1)
users['age'] = pd.to_numeric(users['age'])
print(users.head())
```

### Step 4. Discover what is the mean age per occupation

```python
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print(mean_age_per_occupation)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
gender_counts = users.groupby(['occupation', 'gender']).size().unstack(fill_value=0)
total_per_occupation = users.groupby('occupation').size()
male_ratio = (gender_counts['M'] / total_per_occupation * 100).sort_values(ascending=False)
print(male_ratio)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print(min_max_age_per_occupation)
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age_by_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()
print(mean_age_by_occupation_gender)
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_occupation_counts = users.groupby(['occupation', 'gender']).size().unstack(fill_value=0)
total_by_occupation = gender_occupation_counts.sum(axis=1)
gender_percentage = (gender_occupation_counts.div(total_by_occupation, axis=0) * 100)
print(gender_percentage)
```
