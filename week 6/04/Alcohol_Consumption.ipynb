### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/pandas/drinks.csv).

### Step 3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/drinks.csv')
display(drinks.head())
```

### Step 4. Which continent drinks more beer on average?

```python
beer_consumption_per_continent = drinks.groupby('continent')['beer_servings'].mean()
display(beer_consumption_per_continent)
print(f"Continent that drinks more beer on average: {beer_consumption_per_continent.idxmax()}")
```

### Step 5. For each continent print the statistics for wine consumption.

```python
wine_consumption_stats = drinks.groupby('continent')['wine_servings'].describe()
display(wine_consumption_stats)
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
mean_alcohol_consumption = drinks.groupby('continent').mean(numeric_only=True)
display(mean_alcohol_consumption)
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
median_alcohol_consumption = drinks.groupby('continent').median(numeric_only=True)
display(median_alcohol_consumption)
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
spirit_consumption_stats = drinks.groupby('continent')['spirit_servings'].agg(['mean', 'min', 'max'])
display(spirit_consumption_stats)
```
