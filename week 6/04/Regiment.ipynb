### Step 3. Assign it to a variable called regiment.
#### Don't forget to name each column

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],         'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],         'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],         'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
regiment = pd.DataFrame(raw_data)
```

### Step 4. What is the mean preTestScore from the regiment Nighthawks?  

```python
nighthawks_pretest_mean = regiment[regiment['regiment'] == 'Nighthawks']['preTestScore'].mean()
print(f"Mean preTestScore for Nighthawks regiment: {nighthawks_pretest_mean}")
```

### Step 5. Present general statistics by company

```python
company_stats = regiment.groupby('company').describe()
print("General statistics by company:")
print(company_stats)
```

### Step 6. What is the mean of each company's preTestScore?

```python
company_pretest_mean = regiment.groupby('company')['preTestScore'].mean()
print("Mean preTestScore by company:")
print(company_pretest_mean)
```

### Step 7. Present the mean preTestScores grouped by regiment and company

```python
regiment_company_pretest_mean = regiment.groupby(['regiment', 'company'])['preTestScore'].mean()
print("Mean preTestScores grouped by regiment and company:")
print(regiment_company_pretest_mean)
```

### Step 8. Present the mean preTestScores grouped by regiment and company without heirarchical indexing

```python
regiment_company_pretest_mean_unstacked = regiment.groupby(['regiment', 'company'])['preTestScore'].mean().unstack()
print("Mean preTestScores grouped by regiment and company (without hierarchical indexing):")
print(regiment_company_pretest_mean_unstacked)
```

### Step 9. Group the entire dataframe by regiment and company

```python
grouped_regiment_company = regiment.groupby(['regiment', 'company'])
print("DataFrame grouped by regiment and company:")
print(grouped_regiment_company)
```

### Step 10. What is the number of observations in each regiment and company

```python
observations_count = grouped_regiment_company.size()
print("Number of observations in each regiment and company:")
print(observations_count)
```

### Step 11. Iterate over a group and print the name and the whole data from the regiment

```python
for name, group in grouped_regiment_company:
    print(f"\nRegiment and Company: {name}")
    print(group)
```
