# Visualizing the Titanic Disaster

### Introduction:

This exercise is based on the titanic Disaster dataset avaiable at [Kaggle](https://www.kaggle.com/c/titanic).  
To know more about the variables check [here](https://www.kaggle.com/c/titanic/data)


### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv)

### Step 3. Assign it to a variable titanic

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv')
```

### Step 4. Set PassengerId as the index

```python
titanic = titanic.set_index('PassengerId')
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
sex_counts = titanic['Sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Male/Female Proportion on Titanic')
plt.ylabel('')
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic, x='Age', y='Fare', hue='Sex', palette={'male': 'skyblue', 'female': 'lightcoral'}, alpha=0.7)
plt.title('Fare vs. Age, Colored by Sex')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

### Step 7. How many people survived?

```python
survived_count = titanic['Survived'].sum()
print(f"Number of people who survived: {survived_count}")
```

### Step 8. Create a histogram with the Fare payed

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic['Fare'], bins=30, kde=True)
plt.title('Distribution of Fare Paid')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

### BONUS: Create your own question and answer it.

```python
survival_rate_by_pclass = titanic.groupby('Pclass')['Survived'].mean().reset_index()
print("Survival Rate by Passenger Class:")
print(survival_rate_by_pclass)

plt.figure(figsize=(8, 5))
sns.barplot(data=survival_rate_by_pclass, x='Pclass', y='Survived', hue='Pclass', palette='viridis', legend=False)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```
