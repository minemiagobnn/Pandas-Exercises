### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv')
```

### Step 4. Delete the Unnamed 0 column

```python
tips = tips.drop(columns=['Unnamed: 0'])
```

### Step 5. Plot the total_bill column histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'], kde=True)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Relationship between Total Bill and Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', size='size', data=tips, sizes=(20, 400))
plt.title('Relationship between Total Bill, Tip, and Party Size')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.legend(title='Party Size')
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips, order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Total Bill per Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tip', y='day', hue='sex', data=tips)
plt.title('Tip Amount per Day, Differentiated by Sex')
plt.xlabel('Tip ($)')
plt.ylabel('Day of the Week')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='day', y='total_bill', hue='time', data=tips, order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Total Bill per Day Differentiated by Time')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.legend(title='Time of Day')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
sns.histplot(tips[tips['time'] == 'Dinner']['tip'], kde=True, ax=axes[0])
axes[0].set_title('Tip Distribution for Dinner')
axes[0].set_xlabel('Tip ($)')
axes[0].set_ylabel('Frequency')
sns.histplot(tips[tips['time'] == 'Lunch']['tip'], kde=True, ax=axes[1])
axes[1].set_title('Tip Distribution for Lunch')
axes[1].set_xlabel('Tip ($)')
axes[1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=tips[tips['sex'] == 'Male'], ax=axes[0], palette='viridis')
axes[0].set_title('Total Bill vs. Tip for Male Customers')
axes[0].set_xlabel('Total Bill ($)')
axes[0].set_ylabel('Tip ($)')

sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=tips[tips['sex'] == 'Female'], ax=axes[1], palette='viridis')
axes[1].set_title('Total Bill vs. Tip for Female Customers')
axes[1].set_xlabel('Total Bill ($)')
axes[1].set_ylabel('Tip ($)')

plt.tight_layout()
plt.show()
```
