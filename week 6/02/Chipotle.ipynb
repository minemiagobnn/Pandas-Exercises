### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
url = "https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')
display(chipo.head())
```

### Step 4. How many products cost more than $10.00?

```python

chipo['item_price'] = chipo['item_price'].str.replace('$', '', regex=False).astype(float)

chipo['price_per_item'] = chipo['item_price'] / chipo['quantity']

products_over_10 = chipo[chipo['price_per_item'] > 10.00]

num_products_over_10 = products_over_10['item_name'].nunique()

print(f"Number of products costing more than $10.00: {num_products_over_10}")
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python

item_prices = chipo[['item_name', 'item_price']]

unique_item_prices = item_prices.drop_duplicates(subset=['item_name'])

display(unique_item_prices)
```

### Step 6. Sort by the name of the item

```python

sorted_item_prices = unique_item_prices.sort_values(by='item_name')

display(sorted_item_prices)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
max_price = chipo['item_price'].max()

most_expensive_items = chipo[chipo['item_price'] == max_price]

quantity_most_expensive = most_expensive_items['quantity'].values[0]

print(f"The quantity of the most expensive item ordered was: {quantity_most_expensive}")
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie_salad_bowls = chipo[chipo['item_name'] == 'Veggie Salad Bowl']

num_veggie_salad_bowls_ordered = len(veggie_salad_bowls)

print(f"'Veggie Salad Bowl' was ordered {num_veggie_salad_bowls_ordered} times.")
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
canned_soda_orders = chipo[chipo['item_name'] == 'Canned Soda']

canned_soda_multiple = canned_soda_orders[canned_soda_orders['quantity'] > 1]

num_canned_soda_multiple = len(canned_soda_multiple)

print(f"Someone ordered more than one 'Canned Soda' {num_canned_soda_multiple} times.")
```
