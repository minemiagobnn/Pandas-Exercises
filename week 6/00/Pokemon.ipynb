### Step 4. Ops...it seems the DataFrame columns are in alphabetical order. Place  the order of the columns as name, type, hp, evolution, pokedex

```python
cols_to_front = ['Name', 'Type 1', 'HP']

cols_to_front = [col for col in cols_to_front if col in pokemon.columns]

remaining_cols = [col for col in pokemon.columns if col not in cols_to_front]
pokemon = pokemon[cols_to_front + remaining_cols]
display(pokemon.head())
```

### Step 5. Add another column called place, and insert what you have in mind.

```python
places = ['forest', 'volcano', 'river', 'forest']
pokemon.loc[:, 'place'] = places * (len(pokemon) // len(places))
display(pokemon.head())
```
