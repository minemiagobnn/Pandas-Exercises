### Step 5. Join the two dataframes along columns and assing to all_data_col

```python
import pandas as pd

raw_data_1 = {        'subject_id': ['1', '2', '3', '4', '5'],        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],         'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
raw_data_2 = {        'subject_id': ['4', '5', '6', '7', '8'],        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],         'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

data1 = pd.DataFrame(raw_data_1)
data2 = pd.DataFrame(raw_data_2)

all_data_col = pd.concat([data1, data2], axis=1)
print("all_data_col (joined along columns):")
display(all_data_col)
```

### Step 7. Merge all_data and data3 along the subject_id value

```python
import pandas as pd

raw_data_1 = {        'subject_id': ['1', '2', '3', '4', '5'],        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],         'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
raw_data_2 = {        'subject_id': ['4', '5', '6', '7', '8'],        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],         'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
data1 = pd.DataFrame(raw_data_1)
data2 = pd.DataFrame(raw_data_2)

all_data = pd.concat([data1, data2], ignore_index=True)

raw_data_3 = {        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
data3 = pd.DataFrame(raw_data_3)

merged_data = pd.merge(all_data, data3, on='subject_id')
print("Merged data (all_data and data3 on subject_id):")
display(merged_data)
```

### Step 8. Merge only the data that has the same 'subject_id' on both data1 and data2

```python
import pandas as pd

raw_data_1 = {        'subject_id': ['1', '2', '3', '4', '5'],        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],         'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
raw_data_2 = {        'subject_id': ['4', '5', '6', '7', '8'],        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],         'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
data1 = pd.DataFrame(raw_data_1)
data2 = pd.DataFrame(raw_data_2)

merged_common_id = pd.merge(data1, data2, on='subject_id', how='inner')
print("Merged data (only common subject_id from data1 and data2):")
display(merged_common_id)
```

### Step 9. Merge all values in data1 and data2, with matching records from both sides where available.

```python
import pandas as pd

raw_data_1 = {        'subject_id': ['1', '2', '3', '4', '5'],        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],         'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
raw_data_2 = {        'subject_id': ['4', '5', '6', '7', '8'],        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],         'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
data1 = pd.DataFrame(raw_data_1)
data2 = pd.DataFrame(raw_data_2)

merged_all_values = pd.merge(data1, data2, on='subject_id', how='outer')
print("Merged data (all values from data1 and data2, matching where available):")
display(merged_all_values)
```
