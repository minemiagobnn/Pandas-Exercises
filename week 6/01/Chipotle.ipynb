### Step 9. Which was the most-ordered item?

```python
most_ordered_item = chipo.groupby('item_name')['quantity'].sum().idxmax()
print(most_ordered_item)
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
total_quantity_most_ordered = chipo.groupby('item_name')['quantity'].sum().loc['Chicken Bowl']
print(total_quantity_most_ordered)
```

### Step 11. What was the most ordered item in the choice_description column?

```python
most_ordered_choice = chipo.groupby('choice_description')['quantity'].sum().idxmax()
print(most_ordered_choice)
```

### Step 12. How many items were orderd in total?

```python
total_items_ordered = chipo['quantity'].sum()
print(total_items_ordered)
```

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
print(chipo['item_price'].dtype)
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
```

#### Step 13.c. Check the item price type

```python
print(chipo['item_price'].dtype)
```

### Step 14. How much was the revenue for the period in the dataset?

```python
total_revenue = chipo['item_price'].sum()
print(f"Total revenue for the period: ${total_revenue:.2f}")
```

### Step 15. How many orders were made in the period?

```python
num_orders = chipo['order_id'].nunique()
print(f"Number of orders made: {num_orders}")
```

### Step 16. What is the average revenue amount per order?

```python
average_revenue_per_order = total_revenue / num_orders
print(f"Average revenue per order: ${average_revenue_per_order:.2f}")
```
